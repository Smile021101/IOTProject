```python
import numpy as np
import pandas as pd
```

```python
ph_1=[np.random.randint(67,80)/10 for i in range(2500)]
ph_0=[np.random.randint(0,67)/10 for i in range(2000)]+[np.random.randint(84,140)/10 for i in range(500)]
```

```python
ph_0
```

```python
Turbidity_1=[np.random.randint(415,500)/100 for i in range(2500)]
Turbidity_0=[np.random.randint(0,420)/100 for i in range(2500)]
Turbidity_0
```

```python
Solids_1=[np.random.randint(0,200) for i in range(2500)]
Solids_0=[np.random.randint(170,5000) for i in range(2500)]
```

```python
data_1=pd.DataFrame(ph_1,columns=['PH'])
```

```python
data_1['Solids']=Solids_1
data_1['Turbidity']=Turbidity_1
data_1['Potability']=1
```

```python
data_1
```

```python
data_1.duplicated().sum()
```

```python
data_0=pd.DataFrame(ph_0,columns=['PH'])
data_0['Solids']=Solids_0
data_0['Turbidity']=Turbidity_0
data_0['Potability']=0
```

```python
data_0
```

```python
data_0.duplicated().sum()
```

```python
data=pd.concat([data_0,data_1],ignore_index=True)
```

```python
data
```

```python
data
```

```python
data.to_csv('Water_Portability_IoT.csv')
```
